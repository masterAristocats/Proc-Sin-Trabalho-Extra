```python
import numpy as np

def generate_ula(M, d):
    """
    Gera as coordenadas tridimensionais de um
    Uniform Linear Array (ULA).

    Parâmetros
    ----------
    M : int
        Número de sensores.
    d : float
        Espaçamento entre sensores.

    Retorna
    -------
    positions : ndarray (M, 3)
        Coordenadas (x, y, z) dos sensores.
    """

    if M <= 0:
        raise ValueError("M deve ser maior que zero.")

    if d <= 0:
        raise ValueError("d deve ser maior que zero.")

    positions = np.zeros((M, 3), dtype=float)

    # Sensores distribuídos ao longo do eixo x
    positions[:, 0] = np.arange(M) * d

    return positions
```

###TESTE

```python
M = 8
d = 0.5

positions = generate_ula(M, d)

print(positions)
```

```python
import matplotlib.pyplot as plt

M = 8
d = 0.5

positions = generate_ula(M, d)

plt.figure(figsize=(8,2))

plt.scatter(positions[:,0], positions[:,1], s=80)

# Numeração dos sensores
for i in range(M):
    plt.text(
        positions[i,0],
        positions[i,1] + 0.03,
        str(i),
        ha='center'
    )

plt.title("Uniform Linear Array (ULA)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis("equal")

plt.show()
```

```python
distancias = np.diff(positions[:,0])

print(distancias)
```
