```python
import numpy as np

def beampattern(positions, azimuth, elevation, wavelength, weights=None):
    """
    Calcula o beampattern normalizado em dB.

    Parâmetros
    ----------
    positions : ndarray (M,3)
        Coordenadas dos sensores.

    azimuth : array_like
        Azimutes (rad).

    elevation : float ou array_like
        Elevação (rad).

    wavelength : float
        Comprimento de onda.

    weights : ndarray (M,), opcional
        Vetor de pesos do beamformer. Se None,
        utiliza pesos uniformes.

    Retorna
    -------
    B_dB : ndarray
        Beampattern normalizado em dB.
    """

    M = positions.shape[0]

    # Se nenhum peso for informado, utiliza pesos uniformes
    if weights is None:
        weights = np.ones(M, dtype=complex)

    AF = np.zeros(len(azimuth), dtype=complex)

    for i, phi in enumerate(azimuth):

        a = steering_vector(
            positions,
            phi,
            elevation,
            wavelength
        )

        # Array Factor
        AF[i] = np.conj(weights) @ a

    # Ganho
    B = np.abs(AF)

    # Normalização
    B /= np.max(B)

    # Conversão para dB
    B_dB = 20 * np.log10(np.maximum(B, 1e-12))

    return B_dB
```
