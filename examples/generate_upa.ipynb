```python
from src.generate_upa import generate_upa
import numpy as np
```

```python
Mx = 4
My = 3
dx = 0.5
dy = 0.5

positions = generate_upa(Mx, My, dx, dy)

print(positions)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

plt.scatter(positions[:,0], positions[:,1], s=80)

for i in range(len(positions)):
    plt.text(
        positions[i,0],
        positions[i,1] + 0.03,
        str(i),
        ha="center",
        fontsize=9
    )

plt.title("Uniform Planar Array (UPA)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis("equal")

plt.show()
```
