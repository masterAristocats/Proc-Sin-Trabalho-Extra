```python
from src.generate_ucya import generate_ucya
import numpy as np
```

```python
Mc = 8
Nz = 3
R = 1
dz = 0.5

positions = generate_ucya(Mc, Nz, R, dz)

print(positions)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    positions[:,0],
    positions[:,1],
    positions[:,2],
    s=60
)

for i in range(len(positions)):
    ax.text(
        positions[i,0],
        positions[i,1],
        positions[i,2],
        str(i),
        fontsize=8
    )

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.set_title("Uniform Cylindrical Array (UCYA)")

plt.show()
```
