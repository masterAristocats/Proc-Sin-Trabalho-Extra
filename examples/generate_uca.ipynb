# Arranjo Circular Uniforme (UCA)

## Modelagem Matemática

No Arranjo Circular Uniforme (UCA), os **M sensores** são distribuídos uniformemente sobre uma circunferência de raio **R**, localizada no plano \(xy\).

O ângulo correspondente ao sensor \(m\) é dado por

\[
\phi_m=\frac{2\pi m}{M},
\qquad
m=0,1,\ldots,M-1.
\]

As coordenadas tridimensionais de cada sensor são

\[
\mathbf r_m=
\begin{bmatrix}
R\cos(\phi_m)\\
R\sin(\phi_m)\\
0
\end{bmatrix}.
\]

Dessa forma, todos os sensores permanecem no plano \(z=0\), distribuídos igualmente ao longo da circunferência.

A função `generate_uca(M, R)` retorna uma matriz de dimensão `(M,3)` contendo as coordenadas tridimensionais dos sensores.
